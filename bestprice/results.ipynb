## -- Scenario 1: Brand=1, Category=1, Subcategory=1

```python
df_recs_1 = pd.read_pickle('data/recommendations/df_recos_simple.pkl')
```

```python
df_dscr_1 = df_recs_1.describe()
df_dscr_1 = df_dscr_1.round(2)
df_dscr_1
```

```python
df_cats_1 = df_recs_1.groupby(['Category2'], as_index=True).agg({'N_Brands':'mean',
                                                                 'N_Categories':'mean',
                                                                 'N_Subcategories':'mean',
                                                                 })

df_cats_1 = df_cats_1.round(2)
df_cats_1
```

## - Create table for evaluation

```python
df_dscrs = [df_dscr_1, df_dscr_2,df_dscr_3,df_dscr_4,df_dscr_5,df_dscr_6,df_dscr_7,df_dscr_8,df_dscr_9,df_dscr_10,df_dscr_11]
```

```python
df_rows = []
scenarios = []
i=1
for df_dscr in df_dscrs:
    df_rows.append(df_dscr.iloc[[1]])
    scenarios.append('Scenario '+str(i))
    i += 1
```

```python
df_means = pd.concat(df_rows, axis=0)
df_means = df_means.drop(['doc_id'], axis=1)
df_means['Scenarios'] = scenarios
df_means = df_means[['Scenarios', 'N_Brands', 'N_Categories', 'N_Subcategories']]
df_means
```
